```python
#e) Crear una nueva columna que indique si una película es "Corta" (duración menor a 1.5 horas), 
# "Normal" (entre 1.5 y 2.5 horas) o "Larga" (mayor a 2.5 horas)
condiciones=[movies['length']<90,
             movies['length']<150,
             movies['length']>150]
valores=['Corta','Normal','Larga']

movies['Duración']=np.select(condiciones, valores, default='Invalido')
movies
```

```python
#d) Calcula la mediana de la población de los municipios, y clasifícalos, aquellos por arriba de la mediana deben llamarse “muy poblados” 
# y aquellos debajo de la mediana “poco poblados”

mediana=Asgard['Población'].median()

conditions=[Asgard['Población']<mediana,
             Asgard['Población']>mediana]
values=['Poco Poblado','Muy Poblado']

Asgard['EstatusPoblacional']=np.select(conditions, values, default='Invalido')
Asgard
```
